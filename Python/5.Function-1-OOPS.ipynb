age=int(input("the age is"))

```python
def ageClass():
    if(age<18):
        print("young")
        catg="young"
    else:
        print("old")
        catg="old"
    return catg
```

```python
ageClass()
```

```python
ageCateg=ageClass()
```

```python
ageCateg
```

```python
if (ageCateg=="young"):
    print("selected")
else:
    print("rejected")
```

```python
age=int(input("the age is"))
```

```python
def ageClassNOR():
    if(age<18):
        print("young")
        catg="young"
    else:
        print("old")
        catg="old"
```

```python
ageClassNOR()
```

```python
ageCateg=ageClassNOR()
```

```python
print(ageCateg)
```

```python
if (ageCateg=="young"):
    print("selected")
#ELSE is removed since NON is also NOT young
```
