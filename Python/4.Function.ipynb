```python
#Function-Age classification from user input
```

```python
def ageClass():
    age=int(input("the age:"))
    if (age<18):
        print("young")
    else:
        print("old")
    return age
```

```python
ageClass()
```

```python
#Function-Age classification from lists
```

```python
List=[10,20,30,40,4]
```

```python
def ageClass():
    for age in List:
        if(age<18):
            print("young")
        else:
            print("old")
```

```python
ageClass()
```
