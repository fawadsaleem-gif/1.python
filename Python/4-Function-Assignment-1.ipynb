```python
#Create a function that checks whether the giver number is Odd or Even
```

```python
def OddEven():
    num=int(input("Enter the number:"))
    if(num%2==0):
        print(num,"is Even number")
        message="is Even number"
    else:
        print(num,"is Odd number")
        message="is Odd number"
        return message
```

```python
#Create a funtion that tells eligibilty of marriage for male and female according to their age limit like 21 for male and 18 for female
```

```python
def eligible():
    gender=input("your gender")
    age=int(input("your age"))
    if gender=="Male" and age>21:
        print("Eligible")
        message= "Eligible"
    elif gender=="Female" and age>18:
        print("Eligible")
        message= "Eligible"
    else:
        print("Not Eligible")
        message= "Not Eligible"
    return message
```

```python
marriage=eligible()
```

```python
if (marriage=="Eligible"):
    print("Proceed")
else:
    print("Ignore")
```

```python
def triangle():
    Height=int(input("Height:"))
    Breadth=int(input("Breadth:"))
    Area = (Height*Breadth)/2
    print("Area formula: (Height*Breadth)/2 ")
    print("Area of triangle:", Area)
    
    H1=int(input("Height1:"))
    H2=int(input("Height2:"))
    B1=int(input("Breadth:"))
    Perimeter = H1+H2+B1
    print("Perimeter formula: Height1+Height2+Breadth ")
    print("Perimeter of triangle:",Perimeter)
    return Area, Perimeter
```
